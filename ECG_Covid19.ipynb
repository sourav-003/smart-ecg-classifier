```python
# Preprocessing part

def load_and_preprocess_images(image_folder):
    image_files = os.listdir(image_folder)
    images = []
    for image_file in image_files:
        image_path = os.path.join(image_folder, image_file)
        image = cv2.imread(image_path)
        if image is not None:
            image = cv2.resize(image,(128,128))
            images.append(image)
    return images
```

# EfficientNet (Transfer Learning for ECG images)

```python
import tensorflow as tf
tf.keras.backend.clear_session()

# Force EfficientNet to always use 3 channels
from tensorflow.keras import backend as K
K.set_image_data_format('channels_last')

import os
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.applications.efficientnet import preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# ---------------- Dataset ----------------
dataset_path_train = r"C:\Users\soura\Downloads\smartECG\train"

# Force every image to 3-channel RGB + EfficientNet preprocessing
def preprocess_rgb(img):
    img = np.array(img)
    if len(img.shape) == 2 or img.shape[-1] == 1:  # grayscale → RGB
        img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
    return preprocess_input(img)

# Image parameters
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_rgb,
    validation_split=0.2,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
)

train_generator = train_datagen.flow_from_directory(
    dataset_path_train,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    color_mode="rgb",  
    subset="training"
)

val_generator = train_datagen.flow_from_directory(
    dataset_path_train,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    color_mode="rgb",   
    subset="validation"
)

# ---------------- Model ----------------
base_model = EfficientNetB0(
    weights="imagenet",         
    include_top=False,
    input_shape=(224, 224, 3)  
)

# Freeze base for transfer learning
base_model.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.4)(x)
preds = Dense(train_generator.num_classes, activation="softmax")(x)

model = Model(inputs=base_model.input, outputs=preds)

# Compile model
model.compile(
    optimizer=Adam(learning_rate=1e-3),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

# ---------------- Callbacks ----------------
callbacks = [
    EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True),
    ReduceLROnPlateau(monitor="val_loss", factor=0.2, patience=3, min_lr=1e-6)
]

# ---------------- Training ----------------
print("Starting initial training...")
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=50,
    callbacks=callbacks
)

# ---------------- Fine-tuning ----------------
print("Fine-tuning with base model unfrozen...")
base_model.trainable = True

model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history_finetune = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10,
    callbacks=callbacks
)

# ---------------- Save Model ----------------
save_path = "ecg_efficientnetb0.h5"
model.save(save_path)
print(f"Model saved at {save_path}")
```

```python
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import cv2

# ---------------- Grad-CAM Function ----------------
def make_gradcam_heatmap(img_array, model, last_conv_layer_name="top_conv"):
   
    grad_model = tf.keras.models.Model(
        [model.inputs],
        [model.get_layer(last_conv_layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        class_idx = tf.argmax(predictions[0])
        loss = predictions[:, class_idx]

    grads = tape.gradient(loss, conv_outputs)

    # Pool gradients across channels
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    # Multiply pooled grads with conv outputs
    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)

    # Normalize between 0-1
    heatmap = np.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy(), int(class_idx)

def overlay_gradcam(img_path, heatmap, alpha=0.4):
    """Overlay heatmap on original image"""
    img = cv2.imread(img_path)
    img = cv2.resize(img, (224, 224))
    heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

    superimposed_img = cv2.addWeighted(img, 1-alpha, heatmap, alpha, 0)
    return superimposed_img
```

```python
# Example test image
test_img_path = r"C:\Users\soura\Downloads\smartECG\train\Normal\Normal (1).jpg"

# Preprocess image (same as training pipeline)
img = cv2.imread(test_img_path)
img = cv2.resize(img, (224, 224))
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_preprocessed = np.expand_dims(preprocess_input(img_rgb.astype("float32")), axis=0)

# Generate heatmap
heatmap, pred_class_idx = make_gradcam_heatmap(img_preprocessed, model, last_conv_layer_name="top_conv")

# Overlay on original image
gradcam_img = overlay_gradcam(test_img_path, heatmap)

# Show results
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Original ECG")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(heatmap, cmap="jet")
plt.title("Grad-CAM Heatmap")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(gradcam_img, cv2.COLOR_BGR2RGB))
plt.title(f"Overlay (Predicted: {train_generator.class_indices})")
plt.axis("off")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np

# Plot Training Curves
def plot_training(history, title="Training Curves"):
    plt.figure(figsize=(12, 5))
    
    # Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Acc')
    plt.plot(history.history['val_accuracy'], label='Val Acc')
    plt.title(title + " - Accuracy")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    
    # Loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title(title + " - Loss")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    
    plt.show()

# Plot initial training curves
plot_training(history, "Initial Training")

# Plot fine-tuning training curves
plot_training(history_finetune, "Fine-tuning Training")

# Confusion Matrix
val_generator.reset()
y_true = val_generator.classes
y_pred_probs = model.predict(val_generator)
y_pred = np.argmax(y_pred_probs, axis=1)

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
class_labels = list(val_generator.class_indices.keys())

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()

# Classification report
print("Classification Report:\n", classification_report(y_true, y_pred, target_names=class_labels))

# Sample Predictions
def show_predictions(generator, model, n=5):
    x, y = next(generator)  # one batch
    preds = model.predict(x)
    y_true = np.argmax(y, axis=1)
    y_pred = np.argmax(preds, axis=1)
    
    plt.figure(figsize=(15, 5))
    for i in range(n):
        plt.subplot(1, n, i+1)
        plt.imshow(x[i].astype("uint8"))
        plt.title(f"T:{class_labels[y_true[i]]}\nP:{class_labels[y_pred[i]]}")
        plt.axis("off")
    plt.show()

show_predictions(val_generator, model, n=5)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import keras_tuner as kt
import numpy as np

# Normalize images
X_images = X_images.astype("float32") / 255.0  

num_classes = 5 

def build_model(hp):
    model = keras.Sequential()
    model.add(layers.Conv2D(
        filters=hp.Int('filters', 32, 128, step=32),
        kernel_size=(3, 3),
        activation='relu',
        input_shape=X_images.shape[1:]
    ))
    model.add(layers.MaxPooling2D(pool_size=(2, 2)))
    model.add(layers.Flatten())
    model.add(layers.Dense(
        hp.Int('units', 64, 256, step=64),
        activation='relu'
    ))
    model.add(layers.Dense(num_classes, activation='softmax')) 

    model.compile(
        optimizer=keras.optimizers.Adam(
            hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])
        ),
        loss='sparse_categorical_crossentropy', 
        metrics=['accuracy']
    )
    return model

# Split train/test if not already
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_images, y_images, test_size=0.2, random_state=42, stratify=y_images
)

# Hyperparameter tuning
tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,
    executions_per_trial=1,
    directory='ktuner_ecg',
    project_name='ecg_covid_multiclass'
)

tuner.search(X_train, y_train, epochs=10, validation_data=(X_test, y_test))

# Get best model
best_model = tuner.get_best_models(num_models=1)[0]
best_hps = tuner.get_best_hyperparameters(1)[0]

print("Best hyperparameters:", best_hps.values)

# Evaluate on test
test_loss, test_acc = best_model.evaluate(X_test, y_test, verbose=0)
print("Best model test accuracy:", test_acc)
```

# AutoKeras (Automated Deep Learning)

```python

import autokeras as ak
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical


# If your images are flattened (49153,), reshape them
if len(X_images.shape) == 2:
    # Example reshape: suppose original ECG is (128, 384) grayscale
    H, W = 128, 384
    X_images = X_images.reshape(-1, H, W, 1)

# Normalize pixel values
X_images = X_images.astype("float32") / 255.0

# One-hot encode labels (AutoKeras also accepts raw integers, but one-hot is safe)
y_images = y_images.astype("int")

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_images, y_images, test_size=0.2, random_state=42, stratify=y_images
)

# 2. AutoKeras ImageClassifier
clf = ak.ImageClassifier(
    overwrite=True,
    max_trials=10  # number of different architectures to try
)

# 3. Train
clf.fit(X_train, y_train, epochs=20, validation_split=0.2)

# 4. Evaluate Best Model
loss, acc = clf.evaluate(X_test, y_test)
print("Test Accuracy:", acc)

# 5. Export Best Model
best_model = clf.export_model()
best_model.summary()

# Save model
best_model.save("best_ecg_model")
```
